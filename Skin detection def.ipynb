```python
def skin_detection(image):
    import cv2
    import numpy as np
    import matplotlib.pyplot as plt
    
    image = cv2.imread(image)
    plt.imshow(image)
    
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image)
    
        
    if image.shape[0] > image.shape[1]:
        image_ratio = image.shape[0] / image.shape[1]
        image = cv2.resize(image, (int(200/image_ratio), 200))
    elif image.shape[0] < image.shape[1]:
        image_ratio = image.shape[1] / image.shape[0]
        image = cv2.resize(image, 200, (int(200/image_ratio)))
    
    shape_height = np.array(image.shape)[0]
    shape_width = np.array(image.shape)[1]
    
    
    pallor_African1 = np.array([114, 80, 57]).reshape(1,3)       #1
    pallor_African2 = np.array([210, 162, 128]).reshape(1,3)     #2
    pallor_African3 = np.array([247, 203, 182]).reshape(1,3)     #3

    midtone_African1 = np.array([84, 60, 56]).reshape(1,3)       #4
    midtone_African2 = np.array([159, 106, 92]).reshape(1,3)     #5
    midtone_African3 = np.array([245, 207, 188]).reshape(1,3)    #6

    tanned_African1 = np.array([81, 73, 69]).reshape(1,3)        #7
    tanned_African2 = np.array([127, 107, 89]).reshape(1,3)      #8
    tanned_African3 = np.array([195, 175, 154]).reshape(1,3)     #9

    pallor_Asian1 = np.array([195, 151, 110]).reshape(1,3)       #10
    pallor_Asian2 = np.array([247, 219, 194]).reshape(1,3)       #11
    pallor_Asian3 = np.array([248, 221, 214]).reshape(1,3)       #12

    midtone_Asian1 = np.array([118, 47, 47]).reshape(1,3)        #13
    midtone_Asian2 = np.array([228, 186, 157]).reshape(1,3)      #14
    midtone_Asian3 = np.array([240, 216, 201]).reshape(1,3)      #15

    tanned_Asian1 = np.array([114, 79, 61]).reshape(1,3)         #16
    tanned_Asian2 = np.array([217, 166, 144]).reshape(1,3)       #17
    tanned_Asian3 = np.array([239, 215, 202]).reshape(1,3)       #18

    pallor_caucasian1 = np.array([163, 102, 73]).reshape(1,3)    #19
    pallor_caucasian2 = np.array([233, 189, 156]).reshape(1,3)   #20
    pallor_caucasian3 = np.array([246, 237, 235]).reshape(1,3)   #21

    midtone_caucasian1 = np.array([144, 93, 62]).reshape(1,3)    #22
    midtone_caucasian2 = np.array([187, 138, 156]).reshape(1,3)  #23
    midtone_caucasian3 = np.array([234, 210, 235]).reshape(1,3)  #24

    tanned_caucasian1 = np.array([117, 50, 24]).reshape(1,3)     #25
    tanned_caucasian2 = np.array([204, 151, 105]).reshape(1,3)   #26
    tanned_caucasian3 = np.array([216, 194, 163]).reshape(1,3)   #27

    skin_variations = np.concatenate((pallor_African1, pallor_African2, pallor_African3, midtone_African1, midtone_African2, midtone_African3, tanned_African1, tanned_African2, tanned_African3, pallor_Asian1, pallor_Asian2, pallor_Asian3, midtone_Asian1, midtone_Asian2, midtone_Asian3, tanned_Asian1, tanned_Asian2, tanned_Asian3, pallor_caucasian1, pallor_caucasian2, pallor_caucasian3, midtone_caucasian1, midtone_caucasian2, midtone_caucasian3, tanned_caucasian1, tanned_caucasian2, tanned_caucasian3), axis = 0)

    classification = []
    
    
    
    for row in range(shape_height):
        for column in range(shape_width):
            current_pixel = image[row,column]
            #compare current pixel with skin tones
        
            #iterate through all skin tones color space
            for i in range(skin_variations.shape[0]):
                current_skin_variation_checked_red = skin_variations[i,0]
                if current_pixel[0] in range(current_skin_variation_checked_red - 1, current_skin_variation_checked_red + 1):
                    current_skin_variation_checked_green = skin_variations[i,1]
                    if current_pixel[1] in range(current_skin_variation_checked_green - 1, current_skin_variation_checked_green + 1 ):
                        current_skin_variation_checked_blue = skin_variations[i,2]
                        if current_pixel[2] in range(current_skin_variation_checked_blue - 1, current_skin_variation_checked_blue + 1 ):
                            classification.append(i)

    if len(classification) == 0:
        print("This is an object not a skin")
    else :
        items = {1:"pallor_African", 2:"pallor_African", 3:"pallor_African", 4:"midtone_African", 5:"midtone_African", 6:"midtone_African", 7:"tanned_African", 8:"tanned_African", 9:"tanned_African", 10:"pallor_Asian", 11:"pallor_Asian", 12:"pallor_Asian", 13:"midtone_Asian", 14:"midtone_Asian", 15:"midtone_Asian", 16:"tanned_Asian", 17:"tanned_Asian", 18:"tanned_Asian", 19:"pallor_caucasian", 20:"pallor_caucasian", 21:"pallor_caucasian", 22:"midtone_caucasian", 23:"midtone_caucasian", 24:"midtone_caucasian", 25:"tanned_caucasian", 26:"tanned_caucasian", 27:"tanned_caucasian" }
        highest_key = 0
        highest_value = 0
        for i in range(28):
            i = i+1
            count = classification.count(i)
            if count > highest_value:
                highest_value = count
                highest_key = i
        items.get(highest_key)
        print(f"You are a \"{items.get(highest_key)}\", am i right")
```
